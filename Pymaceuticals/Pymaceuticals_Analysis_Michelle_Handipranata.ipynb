```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem
```

```python
# Hide warning messages in notebook
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load in csv
clinical_trial_csv = pd.read_csv("./Resources/clinicaltrial_data.csv")
mouse_drug_csv = pd.read_csv("./Resources/mouse_drug_data.csv")
```

```python
# Read and print the data
clinical_trial_csv.head()
```

```python
# Read and print the data
mouse_drug_csv.head()
```

```python
# Combine the data into single dataset
merge_data = pd.merge(clinical_trial_csv, mouse_drug_csv, how = "left", on = "Mouse ID")
merge_data.head()
```

### Tumor Response to Treatment

```python
# Calculate the mean tumor volume grouped by Drug and Timepoint
tumor_by_drug_timepoint_mean = merge_data.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].mean()
# tumor_by_drug_timepoint

# Print in DataFrame
tumor_by_drug_timepoint_mean = pd.DataFrame({"Tumor Volume (mm3)": tumor_by_drug_timepoint_mean})

tumor_by_drug_timepoint_mean = tumor_by_drug_timepoint_mean.reset_index()
tumor_by_drug_timepoint_mean.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_by_drug_timepoint_sem = merge_data.groupby(["Drug", "Timepoint"]).sem()["Tumor Volume (mm3)"]

# Print in Data Frame
tumor_by_drug_timepoint_sem = pd.DataFrame(tumor_by_drug_timepoint_sem)
tumor_by_drug_timepoint_sem.head().reset_index()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_by_drug_timepoint_mean = tumor_by_drug_timepoint_mean.reset_index()
tumor_by_drug_timepoint_pivot_mean = tumor_by_drug_timepoint_mean.pivot(index="Timepoint",
                                                                       columns="Drug")["Tumor Volume (mm3)"]

tumor_by_drug_timepoint_sem = tumor_by_drug_timepoint_sem.reset_index()
tumor_by_drug_timepoint_pivot_sem = tumor_by_drug_timepoint_sem.pivot(index="Timepoint",
                                                                     columns="Drug")["Tumor Volume (mm3)"]

# Preview that Reformatting worked
tumor_by_drug_timepoint_pivot_mean
```

```python
# Preview that Reformatting worked
tumor_by_drug_timepoint_pivot_sem
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(tumor_by_drug_timepoint_pivot_mean.index, tumor_by_drug_timepoint_pivot_mean["Capomulin"],
            yerr=tumor_by_drug_timepoint_pivot_sem["Capomulin"],
            color="g", marker="o", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(tumor_by_drug_timepoint_pivot_mean.index, tumor_by_drug_timepoint_pivot_mean["Infubinol"],
            yerr=tumor_by_drug_timepoint_pivot_sem["Infubinol"],
            color="b", marker="^", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(tumor_by_drug_timepoint_pivot_mean.index, tumor_by_drug_timepoint_pivot_mean["Ketapril"],
            yerr=tumor_by_drug_timepoint_pivot_sem["Ketapril"],
            color="r", marker="s", markersize=5, linestyle="dashed", linewidth=0.50)
plt.errorbar(tumor_by_drug_timepoint_pivot_mean.index, tumor_by_drug_timepoint_pivot_mean["Placebo"],
            yerr=tumor_by_drug_timepoint_pivot_sem["Placebo"],
            color="c", marker=".", markersize=5, linestyle="dashed", linewidth=0.50)

plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.grid(axis="y")
plt.legend(["Capomulin", "Infubinol", "Ketapril", "Placebo"],
          loc="best", fontsize="small", fancybox=True)

plt.show()


```

### Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
meta_by_drug_timepoint_mean = merge_data.groupby(["Drug", "Timepoint"])["Metastatic Sites"].mean()

# Convert to DataFrame
meta_by_drug_timepoint_mean = pd.DataFrame(meta_by_drug_timepoint_mean)

# Preview DataFrame
meta_by_drug_timepoint_mean.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
meta_by_drug_timepoint_sem = merge_data.groupby(["Drug", "Timepoint"]).sem()["Metastatic Sites"]

# Convert to DataFrame
meta_by_drug_timepoint_sem = pd.DataFrame(meta_by_drug_timepoint_sem)

# Preview DataFrame
meta_by_drug_timepoint_sem.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
meta_by_drug_timepoint_mean = meta_by_drug_timepoint_mean.reset_index()
meta_by_drug_timepoint_pivot_mean = meta_by_drug_timepoint_mean.pivot(index="Timepoint",
                                                                     columns="Drug")["Metastatic Sites"]

meta_by_drug_timepoint_sem = meta_by_drug_timepoint_sem.reset_index()
meta_by_drug_timepoint_pivot_sem = meta_by_drug_timepoint_sem.pivot(index="Timepoint",
                                                                   columns="Drug")["Metastatic Sites"]

# Preview that Reformatting worked
meta_by_drug_timepoint_pivot_mean
```

```python
# Preview that Reformatting worked
meta_by_drug_timepoint_pivot_sem
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(meta_by_drug_timepoint_pivot_mean.index, meta_by_drug_timepoint_pivot_mean["Capomulin"],
            yerr=meta_by_drug_timepoint_pivot_sem["Capomulin"],
            color="b", marker="^", markersize=5, linestyle="dashed", linewidth=0.5)
plt.errorbar(meta_by_drug_timepoint_pivot_mean.index, meta_by_drug_timepoint_pivot_mean["Infubinol"],
            yerr=meta_by_drug_timepoint_pivot_sem["Infubinol"],
            color="r", marker=".", markersize=5, linestyle="dashed", linewidth=0.5)
plt.errorbar(meta_by_drug_timepoint_pivot_mean.index, meta_by_drug_timepoint_pivot_mean["Ketapril"],
            yerr=meta_by_drug_timepoint_pivot_sem["Ketapril"],
            color="g", marker="o", markersize=5, linestyle="dashed", linewidth=0.5)
plt.errorbar(meta_by_drug_timepoint_pivot_mean.index, meta_by_drug_timepoint_pivot_mean["Placebo"],
            yerr=meta_by_drug_timepoint_pivot_sem["Placebo"],
            color="m", marker="d", markersize=5, linestyle="dashed", linewidth=0.5)


plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Metastatic Sites")
plt.grid(axis="y")
plt.legend(["Capomulin", "Infubinol", "Ketapril", "Placebo"],
          loc="best", fontsize="small", fancybox=True)

# Show the Figure
plt.show()
```

### Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
survival_count = merge_data.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].count()

# Convert to DataFrame
survival_count = pd.DataFrame({"Mouse Count": survival_count})

# Preview DataFrame
survival_count.head().reset_index()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_count = survival_count.reset_index()
survival_count_pivot = survival_count.pivot(index="Timepoint", columns="Drug")["Mouse Count"]

# Preview the Data Frame
survival_count_pivot.head()
```

```python
# Generate the Plot (Accounting for percentages)
plt.plot(100 * survival_count_pivot["Capomulin"] / 25,
        "bo", linestyle="dashed", markersize=5, linewidth=0.50)
plt.plot(100 * survival_count_pivot["Infubinol"] / 25,
        "r+", linestyle="dashed", markersize=5, linewidth=0.50)
plt.plot(100 * survival_count_pivot["Ketapril"] / 25,
        "g.", linestyle="dashed", markersize=5, linewidth=0.50)
plt.plot(100 * survival_count_pivot["Placebo"] / 25,
        "m^", linestyle="dashed", markersize=5, linewidth=0.50)

plt.title("Survival Rates During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Survival Rate (%)")
plt.grid(True)
plt.legend(["Capomulin", "Infubinol", "Ketapril", "Placebo"],
          loc="best", fontsize="small", fancybox=True)

# Show the Figure
plt.show()
```

### Summary Bar Graph

```python
# Calculate the percent changes for each drug
tumor_pct_change = 100 * (tumor_by_drug_timepoint_pivot_mean.iloc[-1] - tumor_by_drug_timepoint_pivot_mean.iloc[0]) / tumor_by_drug_timepoint_pivot_mean.iloc[0]
tumor_pct_sem = 100 * (tumor_by_drug_timepoint_pivot_sem.iloc[-1] - tumor_by_drug_timepoint_pivot_sem.iloc[0]) / tumor_by_drug_timepoint_pivot_sem.iloc[0]

# Display the data to confirm
tumor_pct_change
```

```python
# Store all Relevant Percent Changes into a Tuple
pct_change = (tumor_pct_change["Capomulin"],
             tumor_pct_change["Infubinol"],
             tumor_pct_change["Ketapril"],
             tumor_pct_change["Placebo"])

# Splice the data between passing and failing drugs
fig, ax = plt.subplots()
ind = np.arange(len(pct_change)) 
width = 1
rectsPass = ax.bar(ind[0], pct_change[0], width, color="orange")
rectsFail = ax.bar(ind[1:], pct_change[1:], width, color="blue")

# Orient widths. Add labels, tick marks, etc. 
ax.set_ylabel("% Tumor Volume Change")
ax.set_title("Tumor Change Over 45 Days Treatment")
ax.set_xticks(ind + 0.5)
ax.set_xticklabels(("Capomulin", "Infubinol", "Ketapril", "Placebo"))
ax.set_autoscaley_on(False)
ax.set_ylim([-30, 70])
ax.grid(True)

# Use functions to label the percentages of changes
def autolabelFail(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width() / 2., 3, "%d%%" % int(height),
               ha="center", va="bottom", color="white")

def autolabelPass(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width() / 2., -8, "-%d%%" % int(height),
               ha="center", va="bottom", color="white")
        
# Call functions to implement the function calls
autolabelPass(rectsPass)
autolabelFail(rectsFail)

# Show the Figure
fig.show()
```

### Observation Summary

According to the four treatments we compared (Capomulin, Infubinol, Ketapril, and Placebo):
- Only Capomulin has a continuous decline over time. Infubinol, Ketapril, and Placebo is increasing over time.
- The highest number of mice that still alive over time is Capomulin treatment.
- The bar graph shows negative % tumor volume change for Capomulin, which means tumor is decreasing with Capomulin treatment over time.
